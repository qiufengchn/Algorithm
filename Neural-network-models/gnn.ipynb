```python
# gnn.ipynb
import torch
import torch.nn as nn
import torch.optim as optim

# 1. 定义一个简单的神经网络模型
class SimpleNet(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

# 2. 设置参数并生成随机数据
input_size = 10  # 输入特征数量
hidden_size = 20 # 隐藏层大小
output_size = 1  # 输出大小 (例如，回归问题)
num_samples = 100 # 样本数量
learning_rate = 0.01
num_epochs = 100 # 训练轮数

# 随机生成输入数据和目标标签
# 使用 .float() 确保数据类型是 FloatTensor
X_train = torch.randn(num_samples, input_size).float()
# 假设目标是输入的某种线性组合加上一些噪声
# 这里简化为随机目标，实际应用中应根据问题定义
y_train = torch.randn(num_samples, output_size).float()

# 3. 实例化模型、损失函数和优化器
model = SimpleNet(input_size, hidden_size, output_size)
criterion = nn.MSELoss() # 均方误差损失，适用于回归问题
optimizer = optim.SGD(model.parameters(), lr=learning_rate) # 随机梯度下降

# 4. 训练循环
print("开始训练...")
for epoch in range(num_epochs):
    # 前向传播
    outputs = model(X_train)
    loss = criterion(outputs, y_train)

    # 反向传播和优化
    optimizer.zero_grad() # 清除之前的梯度
    loss.backward()      # 计算梯度
    optimizer.step()     # 更新权重

    # 每隔10轮打印一次损失
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

print("训练完成!")

# (可选) 查看模型的部分预测结果
with torch.no_grad(): # 在评估/预测时不计算梯度
    test_input = torch.randn(1, input_size).float()
    prediction = model(test_input)
    print(f"\n测试输入: {test_input}")
    print(f"模型预测输出: {prediction}")
```

# 神经网络训练过程可视化

为了帮助您更好地理解神经网络训练过程，我可以为您创建一个可视化版本的代码。下面是在原有代码基础上添加了可视化功能的实现：



```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
# 神一样的存在,加进来就解决报错问题!
# 参考链接:https://blog.csdn.net/qq_73968401/article/details/143191306

import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np

# 设置字体为SimHei（黑体）
plt.rcParams['font.sans-serif'] = ['SimHei']
# 解决坐标轴负号显示问题
plt.rcParams['axes.unicode_minus'] = False

# 1. 定义一个简单的神经网络模型
class SimpleNet(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

# 2. 设置参数并生成随机数据
input_size = 2  # 改为2，便于2D可视化
hidden_size = 10 
output_size = 1
num_samples = 100
learning_rate = 0.01
num_epochs = 100

# 创建可视化数据（2D方便可视化）
X_train = torch.randn(num_samples, input_size).float()
# 创建一个非线性目标函数: y = sin(x1) * cos(x2) + 噪声
x1 = X_train[:, 0].numpy()
x2 = X_train[:, 1].numpy()
y_true = np.sin(x1) * np.cos(x2) + np.random.normal(0, 0.1, num_samples)
y_train = torch.FloatTensor(y_true).view(-1, output_size)

# 3. 实例化模型、损失函数和优化器
model = SimpleNet(input_size, hidden_size, output_size)
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

# 存储训练过程中的损失
loss_history = []

# 4. 训练循环
print("开始训练...")
for epoch in range(num_epochs):
    # 前向传播
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    
    # 记录损失
    loss_history.append(loss.item())
    
    # 反向传播和优化
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # 每隔10轮打印一次损失
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

print("训练完成!")

# 5. 可视化训练过程
plt.figure(figsize=(12, 10))

# 5.1 绘制损失函数曲线
plt.subplot(2, 2, 1)
plt.plot(loss_history)
plt.title('损失函数曲线')
plt.xlabel('训练轮数')
plt.ylabel('损失')

# 5.2 可视化原始数据和预测结果（投影到x1轴）
plt.subplot(2, 2, 2)
with torch.no_grad():
    predictions = model(X_train).numpy().flatten()

plt.scatter(x1, y_true, alpha=0.5, label='实际值')
plt.scatter(x1, predictions, alpha=0.5, label='预测值')
plt.title('预测结果与实际值比较 (x1轴投影)')
plt.xlabel('x1')
plt.ylabel('y')
plt.legend()

# 5.3 可视化原始数据和预测结果（投影到x2轴）
plt.subplot(2, 2, 3)
plt.scatter(x2, y_true, alpha=0.5, label='实际值')
plt.scatter(x2, predictions, alpha=0.5, label='预测值')
plt.title('预测结果与实际值比较 (x2轴投影)')
plt.xlabel('x2')
plt.ylabel('y')
plt.legend()

# 5.4 可视化权重矩阵
plt.subplot(2, 2, 4)
weight1 = model.fc1.weight.data.numpy()
plt.imshow(weight1, cmap='viridis')
plt.colorbar()
plt.title('第一层权重矩阵可视化')
plt.xlabel('输入特征')
plt.ylabel('隐藏层节点')

plt.tight_layout()
plt.savefig('neural_network_training_visualization.png')
plt.show()

# 6. 3D可视化输入和预测
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 5))

# 6.1 原始数据的3D图
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(x1, x2, y_true, c=y_true, cmap='viridis')
ax1.set_title('实际数据 3D 可视化')
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('y')

# 6.2 模型预测的3D图
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(x1, x2, predictions, c=predictions, cmap='viridis')
ax2.set_title('模型预测 3D 可视化')
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_zlabel('y')

plt.savefig('neural_network_3d_visualization.png')
plt.tight_layout()
plt.show()
```
