```python
import numpy as np

class QuadTree:
    def __init__(self, boundary, capacity):
        """
        初始化四叉树节点
        boundary: [x, y, width, height] 表示边界
        capacity: 每个节点最多容纳的对象数量
        """
        self.boundary = boundary
        self.capacity = capacity
        self.objects = []  # 存储方块对象
        self.divided = False
        self.northwest = None
        self.northeast = None
        self.southwest = None
        self.southeast = None

    def subdivide(self):
        """将当前节点划分为四个子节点"""
        x, y, w, h = self.boundary
        w_half = w / 2
        h_half = h / 2

        nw = [x, y + h_half, w_half, h_half]
        ne = [x + w_half, y + h_half, w_half, h_half]
        sw = [x, y, w_half, h_half]
        se = [x + w_half, y, w_half, h_half]

        self.northwest = QuadTree(nw, self.capacity)
        self.northeast = QuadTree(ne, self.capacity)
        self.southwest = QuadTree(sw, self.capacity)
        self.southeast = QuadTree(se, self.capacity)
        self.divided = True

    def insert(self, obj_id, position):
        """插入对象到四叉树中"""
        # 检查是否在边界内
        if not self._contains(position):
            return False

        # 如果节点未满且未细分，则直接添加
        if len(self.objects) < self.capacity and not self.divided:
            self.objects.append((obj_id, position))
            return True

        # 如果节点未细分，则进行细分
        if not self.divided:
            self.subdivide()

        # 尝试将对象插入到子节点
        if self.northwest.insert(obj_id, position): return True
        if self.northeast.insert(obj_id, position): return True
        if self.southwest.insert(obj_id, position): return True
        if self.southeast.insert(obj_id, position): return True

        return False

    def query(self, range_rect):
        """查询范围内的所有对象"""
        found = []
        
        # 如果查询范围与当前节点不相交，返回空列表
        if not self._intersects(range_rect):
            return found

        # 检查当前节点中的对象
        for obj_id, pos in self.objects:
            if self._point_in_range(pos, range_rect):
                found.append(obj_id)

        # 如果已细分，递归查询子节点
        if self.divided:
            found.extend(self.northwest.query(range_rect))
            found.extend(self.northeast.query(range_rect))
            found.extend(self.southwest.query(range_rect))
            found.extend(self.southeast.query(range_rect))

        return found

    def _contains(self, point):
        """检查点是否在当前节点边界内"""
        x, y = point
        bx, by, bw, bh = self.boundary
        return (x >= bx and x < bx + bw and 
                y >= by and y < by + bh)

    def _intersects(self, range_rect):
        """检查两个矩形是否相交"""
        rx, ry, rw, rh = range_rect
        bx, by, bw, bh = self.boundary
        return not (rx + rw < bx or
                   rx > bx + bw or
                   ry + rh < by or
                   ry > by + bh)

    def _point_in_range(self, point, range_rect):
        """检查点是否在矩形范围内"""
        x, y = point
        rx, ry, rw, rh = range_rect
        return (x >= rx and x < rx + rw and 
                y >= ry and y < ry + rh)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Rectangle
from matplotlib.collections import LineCollection

class QuadTree:
    """QuadTree implementation for spatial partitioning"""
    def __init__(self, boundary, capacity):
        """
        Initialize QuadTree node
        Args:
            boundary: [x, y, width, height] of the node area
            capacity: maximum number of objects in a node
        """
        self.boundary = boundary
        self.capacity = capacity
        self.objects = []
        self.divided = False
        self.northwest = None
        self.northeast = None
        self.southwest = None
        self.southeast = None

    def subdivide(self):
        """Subdivide current node into four child nodes"""
        x, y, w, h = self.boundary
        w_half = w / 2
        h_half = h / 2

        # Create child nodes
        nw = [x, y + h_half, w_half, h_half]
        ne = [x + w_half, y + h_half, w_half, h_half]
        sw = [x, y, w_half, h_half]
        se = [x + w_half, y, w_half, h_half]

        self.northwest = QuadTree(nw, self.capacity)
        self.northeast = QuadTree(ne, self.capacity)
        self.southwest = QuadTree(sw, self.capacity)
        self.southeast = QuadTree(se, self.capacity)
        self.divided = True

    def insert(self, obj_id, position):
        """Insert object into QuadTree"""
        if not self._contains(position):
            return False

        if len(self.objects) < self.capacity and not self.divided:
            self.objects.append((obj_id, position))
            return True

        if not self.divided:
            self.subdivide()

        return (self.northwest.insert(obj_id, position) or
                self.northeast.insert(obj_id, position) or
                self.southwest.insert(obj_id, position) or
                self.southeast.insert(obj_id, position))

    def query(self, range_rect):
        """Query objects within the given range"""
        found = []
        if not self._intersects(range_rect):
            return found

        for obj_id, pos in self.objects:
            if self._point_in_range(pos, range_rect):
                found.append(obj_id)

        if self.divided:
            found.extend(self.northwest.query(range_rect))
            found.extend(self.northeast.query(range_rect))
            found.extend(self.southwest.query(range_rect))
            found.extend(self.southeast.query(range_rect))

        return found

    def _contains(self, point):
        """Check if point is within node boundary"""
        x, y = point
        bx, by, bw, bh = self.boundary
        return (bx <= x < bx + bw and by <= y < by + bh)

    def _intersects(self, range_rect):
        """Check if range intersects with node boundary"""
        rx, ry, rw, rh = range_rect
        bx, by, bw, bh = self.boundary
        return not (rx + rw < bx or rx > bx + bw or 
                   ry + rh < by or ry > by + bh)

    def _point_in_range(self, point, range_rect):
        """Check if point is within range rectangle"""
        x, y = point
        rx, ry, rw, rh = range_rect
        return (rx <= x < rx + rw and ry <= y < ry + rh)

class CollisionSimulation:
    """2D collision simulation with QuadTree-based spatial partitioning"""
    def __init__(self):
        # Simulation parameters
        self.SPACE_SIZE = 10
        self.OBJECT_SIZE = 1
        self.NUM_OBJECTS = 20
        self.SPEED_RANGE = (-2, 2)
        
        # Initialize random state
        np.random.seed(42)
        
        # Initialize object positions and velocities
        self.positions = np.random.uniform(0, self.SPACE_SIZE-self.OBJECT_SIZE, 
                                         (self.NUM_OBJECTS, 2))
        self.velocities = np.random.uniform(self.SPEED_RANGE[0], 
                                          self.SPEED_RANGE[1], 
                                          (self.NUM_OBJECTS, 2))
        
        # Setup visualization
        self._setup_visualization()

    def _setup_visualization(self):
        """Initialize visualization components"""
        # Create figure and axis
        self.fig, self.ax = plt.subplots(figsize=(10, 10))
        self.objects = []
        
        # Create objects
        for pos in self.positions:
            obj = Rectangle(pos, self.OBJECT_SIZE, self.OBJECT_SIZE,
                          facecolor='blue', alpha=0.5)
            self.objects.append(obj)
            self.ax.add_patch(obj)
        
        # Initialize QuadTree visualization
        self.tree_lines = LineCollection([], colors='gray', 
                                       linestyles='--', alpha=0.5)
        self.ax.add_collection(self.tree_lines)
        
        # Initialize detection range visualization
        self.detection_range = Rectangle((0, 0), 
                                      self.OBJECT_SIZE*3, 
                                      self.OBJECT_SIZE*3,
                                      fill=False, color='red', alpha=0.3)
        self.ax.add_patch(self.detection_range)
        
        # Setup plot properties
        self.ax.set_xlim(-1, self.SPACE_SIZE+1)
        self.ax.set_ylim(-1, self.SPACE_SIZE+1)
        self.ax.set_aspect('equal')
        self.ax.set_title('QuadTree Spatial Partitioning Demo')
        self.ax.grid(True, linestyle=':')
        
        # Add legend
        self.ax.plot([], [], 'b--', alpha=0.5, label='QuadTree Boundaries')
        self.ax.plot([], [], 'rs', alpha=0.3, label='Detection Range')
        self.ax.plot([], [], 'bs', alpha=0.5, label='Moving Objects')
        self.ax.legend()

    def _draw_quadtree(self, node):
        """Draw QuadTree boundaries"""
        lines = []
        if node.divided:
            x, y, w, h = node.boundary
            lines.append([(x + w/2, y), (x + w/2, y + h)])
            lines.append([(x, y + h/2), (x + w, y + h/2)])
            
            for child in [node.northwest, node.northeast, 
                         node.southwest, node.southeast]:
                lines.extend(self._draw_quadtree(child))
        return lines

    def update(self, frame):
        """Update animation frame"""
        # Update positions
        self.positions += self.velocities
        
        # Rebuild QuadTree
        self.quadtree = QuadTree([0, 0, self.SPACE_SIZE, self.SPACE_SIZE], 4)
        for i, pos in enumerate(self.positions):
            self.quadtree.insert(i, pos)
        
        # Update QuadTree visualization
        tree_lines = self._draw_quadtree(self.quadtree)
        self.tree_lines.set_segments(tree_lines)
        
        # Check collisions
        for i in range(self.NUM_OBJECTS):
            # Update detection range
            search_range = [
                self.positions[i,0] - self.OBJECT_SIZE,
                self.positions[i,1] - self.OBJECT_SIZE,
                self.OBJECT_SIZE * 3,
                self.OBJECT_SIZE * 3
            ]
            self.detection_range.set_xy((search_range[0], search_range[1]))
            
            # Handle boundary collisions
            if (self.positions[i,0] <= 0 or 
                self.positions[i,0] >= self.SPACE_SIZE-self.OBJECT_SIZE):
                self.velocities[i,0] *= -1
            if (self.positions[i,1] <= 0 or 
                self.positions[i,1] >= self.SPACE_SIZE-self.OBJECT_SIZE):
                self.velocities[i,1] *= -1
            
            # Keep objects within bounds
            self.positions[i,0] = np.clip(self.positions[i,0], 0, 
                                        self.SPACE_SIZE-self.OBJECT_SIZE)
            self.positions[i,1] = np.clip(self.positions[i,1], 0, 
                                        self.SPACE_SIZE-self.OBJECT_SIZE)
            
            # Update object positions
            self.objects[i].set_xy(self.positions[i])
        
        return self.objects + [self.detection_range, self.tree_lines]
    
    def run(self):
        """Run the simulation"""
        self.anim = FuncAnimation(self.fig, self.update, frames=500,
                                interval=20, blit=True)
        plt.show()

if __name__ == "__main__":
    # Create and run simulation
    simulation = CollisionSimulation()
    simulation.run()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Rectangle
from matplotlib.collections import LineCollection

class QuadTree:
    """四叉树实现"""
    def __init__(self, boundary, capacity):
        self.boundary = boundary
        self.capacity = capacity
        self.objects = []
        self.divided = False
        self.northwest = None
        self.northeast = None
        self.southwest = None
        self.southeast = None

    def subdivide(self):
        x, y, w, h = self.boundary
        w_half = w / 2
        h_half = h / 2

        nw = [x, y + h_half, w_half, h_half]
        ne = [x + w_half, y + h_half, w_half, h_half]
        sw = [x, y, w_half, h_half]
        se = [x + w_half, y, w_half, h_half]

        self.northwest = QuadTree(nw, self.capacity)
        self.northeast = QuadTree(ne, self.capacity)
        self.southwest = QuadTree(sw, self.capacity)
        self.southeast = QuadTree(se, self.capacity)
        self.divided = True

    def insert(self, obj_id, position):
        if not self._contains(position):
            return False

        if len(self.objects) < self.capacity and not self.divided:
            self.objects.append((obj_id, position))
            return True

        if not self.divided:
            self.subdivide()

        if (self.northwest.insert(obj_id, position) or
            self.northeast.insert(obj_id, position) or
            self.southwest.insert(obj_id, position) or
            self.southeast.insert(obj_id, position)):
            return True
        return False

    def query(self, range_rect):
        found = []
        if not self._intersects(range_rect):
            return found

        for obj_id, pos in self.objects:
            if self._point_in_range(pos, range_rect):
                found.append(obj_id)

        if self.divided:
            found.extend(self.northwest.query(range_rect))
            found.extend(self.northeast.query(range_rect))
            found.extend(self.southwest.query(range_rect))
            found.extend(self.southeast.query(range_rect))

        return found

    def _contains(self, point):
        x, y = point
        bx, by, bw, bh = self.boundary
        return (bx <= x < bx + bw and by <= y < by + bh)

    def _intersects(self, range_rect):
        rx, ry, rw, rh = range_rect
        bx, by, bw, bh = self.boundary
        return not (rx + rw < bx or rx > bx + bw or 
                   ry + rh < by or ry > by + bh)

    def _point_in_range(self, point, range_rect):
        x, y = point
        rx, ry, rw, rh = range_rect
        return (rx <= x < rx + rw and ry <= y < ry + rh)

class CollisionSimulation:
    def __init__(self):
        # 基础参数设置
        self.SPACE_SIZE = 10
        self.OBJECT_SIZE = 1
        self.NUM_OBJECTS = 20
        self.SPEED_RANGE = (-1, 1)
        
        np.random.seed(42)
        
        # 初始化位置和速度
        self.positions = np.random.uniform(0, self.SPACE_SIZE-self.OBJECT_SIZE, 
                                         (self.NUM_OBJECTS, 2))
        self.velocities = np.random.uniform(self.SPEED_RANGE[0], 
                                          self.SPEED_RANGE[1], 
                                          (self.NUM_OBJECTS, 2))
        
        # 初始化四叉树
        self.quadtree = QuadTree([0, 0, self.SPACE_SIZE, self.SPACE_SIZE], 4)
        
        # 设置可视化
        self._setup_visualization()

    def _setup_visualization(self):
        self.fig, self.ax = plt.subplots(figsize=(10, 10))
        self.objects = []
        self.object_labels = []
        
        # 创建方块和标签
        for i, pos in enumerate(self.positions):
            obj = Rectangle(pos, self.OBJECT_SIZE, self.OBJECT_SIZE,
                          facecolor='lightblue', edgecolor='blue', alpha=0.6)
            self.objects.append(obj)
            self.ax.add_patch(obj)
            
            label = self.ax.text(pos[0] + self.OBJECT_SIZE/2, 
                               pos[1] + self.OBJECT_SIZE/2,
                               str(i), ha='center', va='center')
            self.object_labels.append(label)
        
        # 四叉树可视化
        self.tree_lines = LineCollection([], colors='gray', 
                                       linestyles='dashed', alpha=0.5)
        self.ax.add_collection(self.tree_lines)
        
        # 检测范围可视化
        self.detection_range = Rectangle((0, 0), 
                                      self.OBJECT_SIZE*3, 
                                      self.OBJECT_SIZE*3,
                                      fill=False, color='red', alpha=0.3)
        self.ax.add_patch(self.detection_range)
        
        # 设置图表
        self.ax.set_xlim(-1, self.SPACE_SIZE+1)
        self.ax.set_ylim(-1, self.SPACE_SIZE+1)
        self.ax.set_aspect('equal')
        self.ax.set_title('Quadtree Spatial Partitioning')
        self.ax.grid(True, linestyle=':')
        
        # 修复图例
        self.ax.plot([], [], color='gray', linestyle='--', alpha=0.5, label='Quadtree')
        self.ax.plot([], [], color='red', linestyle='-', alpha=0.3, label='Detection Range')
        self.ax.plot([], [], color='lightblue', alpha=0.6, label='Objects')
        self.ax.plot([], [], color='red', linestyle=':', alpha=0.8, label='Collision')
        self.ax.legend()

    def _draw_quadtree(self, node):
        lines = []
        if node.divided:
            x, y, w, h = node.boundary
            lines.append([(x + w/2, y), (x + w/2, y + h)])
            lines.append([(x, y + h/2), (x + w, y + h/2)])
            
            for child in [node.northwest, node.northeast, 
                         node.southwest, node.southeast]:
                lines.extend(self._draw_quadtree(child))
        return lines

    def check_overlap(self, pos1, pos2):
        x1, y1 = pos1
        x2, y2 = pos2
        return (abs(x1 - x2) < self.OBJECT_SIZE and 
                abs(y1 - y2) < self.OBJECT_SIZE)

    def update(self, frame):
        # 更新位置
        self.positions += self.velocities
        
        # 重建四叉树
        self.quadtree = QuadTree([0, 0, self.SPACE_SIZE, self.SPACE_SIZE], 4)
        for i, pos in enumerate(self.positions):
            self.quadtree.insert(i, pos)
        
        # 更新四叉树可视化
        tree_lines = self._draw_quadtree(self.quadtree)
        self.tree_lines.set_segments(tree_lines)
        
        # 清除旧的碰撞线
        for line in self.ax.lines[4:]:
            line.remove()
        
        # 检查碰撞
        for i in range(self.NUM_OBJECTS):
            # 更新检测范围
            search_range = [
                self.positions[i,0] - self.OBJECT_SIZE,
                self.positions[i,1] - self.OBJECT_SIZE,
                self.OBJECT_SIZE * 3,
                self.OBJECT_SIZE * 3
            ]
            self.detection_range.set_xy((search_range[0], search_range[1]))
            
            # 查询可能的碰撞
            nearby = self.quadtree.query(search_range)
            
            # 检查实际碰撞
            for j in nearby:
                if i < j:
                    if self.check_overlap(self.positions[i], self.positions[j]):
                        self.ax.plot([self.positions[i,0] + self.OBJECT_SIZE/2,
                                    self.positions[j,0] + self.OBJECT_SIZE/2],
                                   [self.positions[i,1] + self.OBJECT_SIZE/2,
                                    self.positions[j,1] + self.OBJECT_SIZE/2],
                                   color='red', linestyle=':', alpha=0.8)
            
            # 边界碰撞
            if (self.positions[i,0] <= 0 or 
                self.positions[i,0] >= self.SPACE_SIZE-self.OBJECT_SIZE):
                self.velocities[i,0] *= -1
            if (self.positions[i,1] <= 0 or 
                self.positions[i,1] >= self.SPACE_SIZE-self.OBJECT_SIZE):
                self.velocities[i,1] *= -1
            
            # 保持在边界内
            self.positions[i] = np.clip(self.positions[i], 0, 
                                      self.SPACE_SIZE-self.OBJECT_SIZE)
            
            # 更新位置
            self.objects[i].set_xy(self.positions[i])
            self.object_labels[i].set_position((self.positions[i,0] + self.OBJECT_SIZE/2,
                                              self.positions[i,1] + self.OBJECT_SIZE/2))
        
        return self.objects + self.object_labels + [self.detection_range, self.tree_lines]

    def run(self):
        self.anim = FuncAnimation(self.fig, self.update, frames=500,
                                interval=50, blit=True)
        plt.show()

if __name__ == "__main__":
    simulation = CollisionSimulation()
    simulation.run()
```
