```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from collections import defaultdict

class Square:
    def __init__(self, x, y, vx=0, vy=0):
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.width = 1
        self.height = 1

class Grid:
    def __init__(self, width, height, cell_size):
        self.cell_size = cell_size
        self.width = width
        self.height = height
        self.cells = defaultdict(list)
    
    def get_cell_coords(self, x, y):
        return (int(x // self.cell_size), int(y // self.cell_size))
    
    def add_square(self, square, square_id):
        cell_x, cell_y = self.get_cell_coords(square.x, square.y)
        self.cells[(cell_x, cell_y)].append(square_id)
    
    def update_grid(self, squares):
        self.cells.clear()
        for i, square in enumerate(squares):
            self.add_square(square, i)
    
    def check_collision(self, square1, square2):
        return (square1.x < square2.x + square2.width and
                square1.x + square1.width > square2.x and
                square1.y < square2.y + square2.height and
                square1.y + square1.height > square2.y)

    def get_neighbors(self, cell_x, cell_y):
        neighbors = []
        for dx in [-1, 0, 1]:
            for dy in [-1, 0, 1]:
                neighbors.append((cell_x + dx, cell_y + dy))
        return neighbors

    def detect_collisions(self, squares):
        collisions = []
        for cell, square_ids in self.cells.items():
            # 检查同一网格内的碰撞
            for i in range(len(square_ids)):
                for j in range(i + 1, len(square_ids)):
                    if self.check_collision(squares[square_ids[i]], squares[square_ids[j]]):
                        collisions.append((square_ids[i], square_ids[j]))
            
            # 检查相邻网格的碰撞
            for neighbor in self.get_neighbors(*cell):
                if neighbor in self.cells:
                    for id1 in square_ids:
                        for id2 in self.cells[neighbor]:
                            if id1 < id2:  # 避免重复检测
                                if self.check_collision(squares[id1], squares[id2]):
                                    collisions.append((id1, id2))
        return collisions

# 设置随机种子
np.random.seed(42)

# 创建画布
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# 生成方块
squares = []
x_coords = np.random.uniform(0, 9, 20)
y_coords = np.random.uniform(0, 9, 20)
for x, y in zip(x_coords, y_coords):
    squares.append(Square(x, y))

# 创建网格系统 (将10x10的空间分成2x2的网格)
grid = Grid(10, 10, 2)
grid.update_grid(squares)

# 检测碰撞
collisions = grid.detect_collisions(squares)

# 绘制方块和网格
# 绘制网格线
for i in range(0, 11, 2):
    ax.axhline(y=i, color='gray', linestyle=':')
    ax.axvline(x=i, color='gray', linestyle=':')

# 绘制方块
for i, square in enumerate(squares):
    color = 'red' if any(i in pair for pair in collisions) else 'blue'
    rect = Rectangle((square.x, square.y), 1, 1, facecolor=color, alpha=0.5)
    ax.add_patch(rect)
    ax.text(square.x + 0.5, square.y + 0.5, str(i+1),
            horizontalalignment='center',
            verticalalignment='center',
            color='white',
            fontweight='bold')

ax.grid(True)
ax.set_title('Grid-based Collision Detection')

# 打印碰撞信息
print("Detected collisions between squares:", collisions)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle as MplRectangle
import time

# 定义矩形边界类
class Rectangle:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height

    def contains(self, point):
        return (point.x >= self.x and point.x <= self.x + self.width and
                point.y >= self.y and point.y <= self.y + self.height)

    def intersects(self, other):
        return not (other.x >= self.x + self.width or
                   other.x + other.width <= self.x or
                   other.y >= self.y + self.height or
                   other.y + other.height <= self.y)

# 定义方块类
class Square:
    def __init__(self, x, y, vx=0, vy=0):
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.width = 1
        self.height = 1

# 方案一：对象复制到相交的叶子节点
class QuadTreeV1:
    def __init__(self, boundary, capacity):
        self.boundary = boundary
        self.capacity = capacity
        self.points = []
        self.divided = False
        self.northeast = None
        self.northwest = None
        self.southeast = None
        self.southwest = None

    def subdivide(self):
        x = self.boundary.x
        y = self.boundary.y
        w = self.boundary.width / 2
        h = self.boundary.height / 2

        ne = Rectangle(x + w, y + h, w, h)
        nw = Rectangle(x, y + h, w, h)
        se = Rectangle(x + w, y, w, h)
        sw = Rectangle(x, y, w, h)

        self.northeast = QuadTreeV1(ne, self.capacity)
        self.northwest = QuadTreeV1(nw, self.capacity)
        self.southeast = QuadTreeV1(se, self.capacity)
        self.southwest = QuadTreeV1(sw, self.capacity)
        self.divided = True

    def intersects_boundary(self, square):
        return (square.x < self.boundary.x + self.boundary.width and
                square.x + square.width > self.boundary.x and
                square.y < self.boundary.y + self.boundary.height and
                square.y + square.height > self.boundary.y)

    def insert(self, square):
        if not self.intersects_boundary(square):
            return False

        if len(self.points) < self.capacity and not self.divided:
            self.points.append(square)
            return True

        if not self.divided:
            self.subdivide()
            existing_points = self.points
            self.points = []
            for point in existing_points:
                self.insert(point)

        inserted = False
        if self.northeast.intersects_boundary(square):
            inserted = self.northeast.insert(square) or inserted
        if self.northwest.intersects_boundary(square):
            inserted = self.northwest.insert(square) or inserted
        if self.southeast.intersects_boundary(square):
            inserted = self.southeast.insert(square) or inserted
        if self.southwest.intersects_boundary(square):
            inserted = self.southwest.insert(square) or inserted
        return inserted

    def query(self, range_rect, found=None):
        if found is None:
            found = set()
            
        if not range_rect.intersects(self.boundary):
            return found
        
        for point in self.points:
            if range_rect.contains(point):
                found.add(point)
                
        if self.divided:
            self.northwest.query(range_rect, found)
            self.northeast.query(range_rect, found)
            self.southwest.query(range_rect, found)
            self.southeast.query(range_rect, found)
            
        return found

# 方案二：对象保留在合适的层级
class QuadTreeV2:
    def __init__(self, boundary, capacity):
        self.boundary = boundary
        self.capacity = capacity
        self.points = []
        self.divided = False
        self.northeast = None
        self.northwest = None
        self.southeast = None
        self.southwest = None

    def subdivide(self):
        x = self.boundary.x
        y = self.boundary.y
        w = self.boundary.width / 2
        h = self.boundary.height / 2

        ne = Rectangle(x + w, y + h, w, h)
        nw = Rectangle(x, y + h, w, h)
        se = Rectangle(x + w, y, w, h)
        sw = Rectangle(x, y, w, h)

        self.northeast = QuadTreeV2(ne, self.capacity)
        self.northwest = QuadTreeV2(nw, self.capacity)
        self.southeast = QuadTreeV2(se, self.capacity)
        self.southwest = QuadTreeV2(sw, self.capacity)
        self.divided = True

    def can_fully_contain(self, square):
        return (square.x >= self.boundary.x and
                square.x + square.width <= self.boundary.x + self.boundary.width and
                square.y >= self.boundary.y and
                square.y + square.height <= self.boundary.y + self.boundary.height)

    def get_containing_child(self, square):
        if not self.divided:
            return None
            
        if self.northeast.can_fully_contain(square):
            return self.northeast
        if self.northwest.can_fully_contain(square):
            return self.northwest
        if self.southeast.can_fully_contain(square):
            return self.southeast
        if self.southwest.can_fully_contain(square):
            return self.southwest
        return None

    def intersects_boundary(self, square):
        return (square.x < self.boundary.x + self.boundary.width and
                square.x + square.width > self.boundary.x and
                square.y < self.boundary.y + self.boundary.height and
                square.y + square.height > self.boundary.y)

    def insert(self, square):
        if not self.intersects_boundary(square):
            return False

        if len(self.points) < self.capacity:
            self.points.append(square)
            return True

        if not self.divided:
            self.subdivide()

        containing_child = self.get_containing_child(square)
        if containing_child:
            return containing_child.insert(square)
        else:
            self.points.append(square)
            return True

    def query(self, range_rect, found=None):
        if found is None:
            found = set()
            
        if not range_rect.intersects(self.boundary):
            return found
        
        for point in self.points:
            if range_rect.contains(point):
                found.add(point)
                
        if self.divided:
            self.northwest.query(range_rect, found)
            self.northeast.query(range_rect, found)
            self.southwest.query(range_rect, found)
            self.southeast.query(range_rect, found)
            
        return found

def draw_tree_structure(qt, ax, squares, title):
    """绘制四叉树的层级结构图"""
    def get_tree_info(node, level=0, x_pos=0):
        info = {
            'level': level,
            'x_pos': x_pos,
            'points': [i+1 for i, s in enumerate(squares) if s in node.points],
            'children': []
        }
        
        if node.divided:
            spacing = 3.0 / (2 ** level)
            children = [
                (node.northwest, x_pos - spacing),
                (node.northeast, x_pos - spacing/3),
                (node.southwest, x_pos + spacing/3),
                (node.southeast, x_pos + spacing)
            ]
            for child, new_x_pos in children:
                info['children'].append(get_tree_info(child, level + 1, new_x_pos))
        
        return info

    def draw_node(info, y_pos, parent_x=None, parent_y=None):
        x_pos = info['x_pos']
        size = 800 if info['points'] else 400
        color = 'lightblue' if info['points'] else 'white'
        
        if parent_x is not None:
            ax.plot([parent_x, x_pos], [parent_y, y_pos], 
                   'gray', alpha=0.5, linestyle='--', zorder=1)
        
        ax.scatter([x_pos], [y_pos], s=size, c=color, 
                  edgecolor='gray', alpha=0.3, zorder=2)
        
        if info['points']:
            ax.text(x_pos, y_pos, f"{info['points']}", 
                   ha='center', va='center',
                   fontsize=9, zorder=3)
        
        for child in info['children']:
            draw_node(child, y_pos - 1.5, x_pos, y_pos)

    tree_info = get_tree_info(qt)
    max_level = max([info['level'] for info in tree_info['children']], default=0) + 1
    
    ax.set_xlim(-6, 6)
    ax.set_ylim(-(max_level*1.5)-2, 2)
    
    draw_node(tree_info, 0)
    
    ax.set_title(title, pad=20)
    ax.axis('off')

def traverse_tree(node):
    """遍历四叉树的所有节点"""
    yield node
    if node.divided:
        yield from traverse_tree(node.northwest)
        yield from traverse_tree(node.northeast)
        yield from traverse_tree(node.southwest)
        yield from traverse_tree(node.southeast)

def draw_quadtree_v1(squares):
    """绘制方案一的四叉树结构"""
    boundary = Rectangle(0, 0, 10, 10)
    qt1 = QuadTreeV1(boundary, 4)
    
    fig = plt.figure(figsize=(24, 16))
    ax = fig.add_subplot(111)
    
    start_time = time.time()
    for square in squares:
        qt1.insert(square)
    insert_time = time.time() - start_time
    
    draw_tree_structure(qt1, ax, squares, 
                       f'Quadtree V1: Copy to Leaves\nInsertion Time: {insert_time:.4f}s')
    
    plt.tight_layout(pad=3.0)
    plt.show()
    
    total_points = sum(len(node.points) for node in traverse_tree(qt1))
    print(f"\n方案一统计信息:")
    print(f"- 插入时间: {insert_time:.4f}秒")
    print(f"- 总方块数量: {len(squares)}")
    print(f"- 树中方块总数: {total_points} (包含重复)")
    print(f"- 平均每个方块被复制次数: {total_points/len(squares):.2f}")

def draw_quadtree_v2(squares):
    """绘制方案二的四叉树结构"""
    boundary = Rectangle(0, 0, 10, 10)
    qt2 = QuadTreeV2(boundary, 4)
    
    fig = plt.figure(figsize=(24, 16))
    ax = fig.add_subplot(111)
    
    start_time = time.time()
    for square in squares:
        qt2.insert(square)
    insert_time = time.time() - start_time
    
    draw_tree_structure(qt2, ax, squares, 
                       f'Quadtree V2: Keep at Appropriate Level\nInsertion Time: {insert_time:.4f}s')
    
    plt.tight_layout(pad=3.0)
    plt.show()
    
    total_points = sum(len(node.points) for node in traverse_tree(qt2))
    print(f"\n方案二统计信息:")
    print(f"- 插入时间: {insert_time:.4f}秒")
    print(f"- 总方块数量: {len(squares)}")
    print(f"- 树中方块总数: {total_points} (无重复)")
    print(f"- 树的节点数量: {len(list(traverse_tree(qt2)))}")

# 生成测试数据
np.random.seed(42)
squares = []
x_coords = np.random.uniform(0, 9, 20)
y_coords = np.random.uniform(0, 9, 20)
for x, y in zip(x_coords, y_coords):
    squares.append(Square(x, y))

# 分别绘制两种方案
print("方案一：将对象复制到相交的叶子节点")
draw_quadtree_v1(squares)

print("\n方案二：将对象保留在合适的层级")
draw_quadtree_v2(squares)
```

```python
def check_collision(s1, s2):
    """检查两个方块是否碰撞"""
    return (s1.x < s2.x + s2.width and
            s1.x + s1.width > s2.x and
            s1.y < s2.y + s2.height and
            s1.y + s1.height > s2.y)

def draw_collision_detection(squares, qt, title, ax):
    """绘制碰撞检测结果"""
    def draw_quadtree_bounds(node):
        # 绘制四叉树边界，使用深色
        rect = MplRectangle((node.boundary.x, node.boundary.y),
                          node.boundary.width, node.boundary.height,
                          fill=False, color='navy', linewidth=1.5, alpha=0.8)
        ax.add_patch(rect)
        
        if node.divided:
            draw_quadtree_bounds(node.northeast)
            draw_quadtree_bounds(node.northwest)
            draw_quadtree_bounds(node.southeast)
            draw_quadtree_bounds(node.southwest)

    # 检测碰撞
    collisions = set()
    for i, square in enumerate(squares):
        query_range = Rectangle(square.x - 1, square.y - 1, 3, 3)
        nearby = qt.query(query_range)
        
        for other in nearby:
            j = squares.index(other)
            if i != j and check_collision(square, other):
                collisions.add(tuple(sorted([i, j])))

    # 绘制四叉树边界
    draw_quadtree_bounds(qt)

    # 在碰撞检测结果中添加查询范围的可视化
    for i, square in enumerate(squares):
        # 绘制查询范围
        query_range = Rectangle(square.x - 1, square.y - 1, 3, 3)
        query_rect = MplRectangle((query_range.x, query_range.y),
                                query_range.width, query_range.height,
                                fill=False, color='yellow', linestyle='--',
                                linewidth=0.5, alpha=0.3)
        ax.add_patch(query_rect)

    # 绘制方块
    for i, square in enumerate(squares):
        is_colliding = any(i in collision for collision in collisions)
        color = 'red' if is_colliding else 'blue'
        rect = MplRectangle((square.x, square.y), 1, 1,
                          facecolor=color, alpha=0.5)
        ax.add_patch(rect)
        ax.text(square.x + 0.5, square.y + 0.5, str(i+1),
                ha='center', va='center',
                color='white', fontweight='bold')

    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.grid(True, linestyle='--', alpha=0.3)
    ax.set_title(f'{title}\nCollisions: {len(collisions)} pairs')

def compare_collision_detection(squares):
    """比较两种方案的碰撞检测结果"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    
    # 方案一
    boundary = Rectangle(0, 0, 10, 10)
    qt1 = QuadTreeV1(boundary, 4)
    for square in squares:
        qt1.insert(square)
    draw_collision_detection(squares, qt1, 'Collision Detection with V1: Copy to Leaves', ax1)
    
    # 方案二
    qt2 = QuadTreeV2(boundary, 4)
    for square in squares:
        qt2.insert(square)
    draw_collision_detection(squares, qt2, 'Collision Detection with V2: Keep at Appropriate Level', ax2)
    
    plt.tight_layout()
    plt.show()

# 运行对比
compare_collision_detection(squares)
```

```python
class LooseQuadTree:
    def __init__(self, boundary, capacity, looseness=1.5):
        """
        初始化松散四叉树
        looseness: 松散因子，边界扩展的比例（默认1.5表示扩展50%）
        """
        self.boundary = boundary
        self.capacity = capacity
        self.points = []
        self.divided = False
        self.northeast = None
        self.northwest = None
        self.southeast = None
        self.southwest = None
        self.looseness = looseness
        
        # 计算松散边界
        self.loose_boundary = self._calculate_loose_boundary()

    def _calculate_loose_boundary(self):
        """计算松散边界"""
        center_x = self.boundary.x + self.boundary.width / 2
        center_y = self.boundary.y + self.boundary.height / 2
        
        loose_width = self.boundary.width * self.looseness
        loose_height = self.boundary.height * self.looseness
        
        loose_x = center_x - loose_width / 2
        loose_y = center_y - loose_height / 2
        
        return Rectangle(loose_x, loose_y, loose_width, loose_height)

    def subdivide(self):
        """分割节点，使用原始边界进行分割"""
        x = self.boundary.x
        y = self.boundary.y
        w = self.boundary.width / 2
        h = self.boundary.height / 2

        ne = Rectangle(x + w, y + h, w, h)
        nw = Rectangle(x, y + h, w, h)
        se = Rectangle(x + w, y, w, h)
        sw = Rectangle(x, y, w, h)

        self.northeast = LooseQuadTree(ne, self.capacity, self.looseness)
        self.northwest = LooseQuadTree(nw, self.capacity, self.looseness)
        self.southeast = LooseQuadTree(se, self.capacity, self.looseness)
        self.southwest = LooseQuadTree(sw, self.capacity, self.looseness)
        self.divided = True

    def insert(self, point):
        """插入点，使用松散边界进行判断"""
        # 检查松散边界
        if not self.loose_boundary.contains(point):
            return False

        if len(self.points) < self.capacity and not self.divided:
            self.points.append(point)
            return True

        if not self.divided:
            self.subdivide()

        # 尝试插入到子节点
        if self.northeast.insert(point): return True
        if self.northwest.insert(point): return True
        if self.southeast.insert(point): return True
        if self.southwest.insert(point): return True

        return False

    def query(self, range_rect, found=None):
        """查询范围内的点，使用松散边界进行判断"""
        if found is None:
            found = set()
            
        if not range_rect.intersects(self.loose_boundary):
            return found
        
        for point in self.points:
            if range_rect.contains(point):
                found.add(point)
                
        if self.divided:
            self.northwest.query(range_rect, found)
            self.northeast.query(range_rect, found)
            self.southwest.query(range_rect, found)
            self.southeast.query(range_rect, found)
            
        return found
```

```python
class HBBNode:
    """层次包围盒节点"""
    def __init__(self, boundary=None):
        self.boundary = boundary  # 当前节点的包围盒
        self.squares = []        # 当前节点包含的方块
        self.children = []       # 子节点
        self.level = 0          # 节点层级

class HBBTree:
    """层次包围盒树"""
    def __init__(self, max_depth=4, min_size=1.0):
        self.root = HBBNode()
        self.max_depth = max_depth
        self.min_size = min_size

    def build(self, squares):
        """构建层次包围盒树"""
        if not squares:
            return
        
        # 计算根节点的包围盒
        min_x = min(s.x for s in squares)
        min_y = min(s.y for s in squares)
        max_x = max(s.x + s.width for s in squares)
        max_y = max(s.y + s.height for s in squares)
        
        self.root.boundary = Rectangle(min_x, min_y, 
                                     max_x - min_x, 
                                     max_y - min_y)
        self.root.squares = squares
        
        # 递归构建树
        self._subdivide(self.root, 0)

    def _subdivide(self, node, depth):
        """递归划分节点"""
        if depth >= self.max_depth or len(node.squares) <= 1:
            return
        
        # 计算当前节点包围盒的中心
        center_x = node.boundary.x + node.boundary.width / 2
        center_y = node.boundary.y + node.boundary.height / 2
        
        # 按四叉树方式划分空间
        quadrants = {
            'nw': [], 'ne': [], 'sw': [], 'se': []
        }
        
        # 将方块分配到合适的象限
        for square in node.squares:
            # 使用方块的中心点进行划分
            square_center_x = square.x + square.width / 2
            square_center_y = square.y + square.height / 2
            
            if square_center_x <= center_x:
                if square_center_y <= center_y:
                    quadrants['sw'].append(square)
                else:
                    quadrants['nw'].append(square)
            else:
                if square_center_y <= center_y:
                    quadrants['se'].append(square)
                else:
                    quadrants['ne'].append(square)
        
        # 为每个非空象限创建子节点
        for direction, squares in quadrants.items():
            if squares:
                child = HBBNode()
                child.squares = squares
                child.level = depth + 1
                
                # 计算子节点的包围盒
                min_x = min(s.x for s in squares)
                min_y = min(s.y for s in squares)
                max_x = max(s.x + s.width for s in squares)
                max_y = max(s.y + s.height for s in squares)
                
                child.boundary = Rectangle(min_x, min_y,
                                        max_x - min_x,
                                        max_y - min_y)
                
                node.children.append(child)
                self._subdivide(child, depth + 1)

def draw_hbb_tree(squares):
    """绘制层次包围盒树的碰撞检测结果"""
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    
    # 构建HBB树
    hbb = HBBTree(max_depth=4)
    
    start_time = time.time()
    hbb.build(squares)
    build_time = time.time() - start_time
    
    def draw_hbb_bounds(node, level=0):
        """递归绘制包围盒"""
        if node.boundary:
            # 根据层级设置不同的颜色
            colors = ['navy', 'green', 'purple', 'orange', 'brown']
            color = colors[level % len(colors)]
            
            rect = MplRectangle((node.boundary.x, node.boundary.y),
                              node.boundary.width, node.boundary.height,
                              fill=False, color=color,
                              linewidth=2.0-level*0.3,
                              alpha=0.8-level*0.1)
            ax.add_patch(rect)
            
            # 显示节点包含的方块数量
            if node.squares:
                ax.text(node.boundary.x + node.boundary.width/2,
                       node.boundary.y + node.boundary.height/2,
                       f'n={len(node.squares)}',
                       ha='center', va='center',
                       color=color, alpha=0.7)
        
        # 递归绘制子节点
        for child in node.children:
            draw_hbb_bounds(child, level+1)
    
    # 绘制HBB树结构
    draw_hbb_bounds(hbb.root)
    
    # 绘制方块
    for i, square in enumerate(squares):
        rect = MplRectangle((square.x, square.y), 1, 1,
                          facecolor='blue', alpha=0.5)
        ax.add_patch(rect)
        ax.text(square.x + 0.5, square.y + 0.5, str(i+1),
                ha='center', va='center',
                color='white', fontweight='bold')

    ax.set_xlim(-1, 11)
    ax.set_ylim(-1, 11)
    ax.grid(True, linestyle='--', alpha=0.3)
    ax.set_title(f'Hierarchical Bounding Box Tree\nBuild Time: {build_time:.4f}s')
    
    plt.tight_layout()
    plt.show()

# 运行层次包围盒树的可视化
print("层次包围盒树方案：")
draw_hbb_tree(squares)
```

```python
import numpy as np

class Vector2D:
    """2D向量类"""
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Vector2D(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        return Vector2D(self.x * scalar, self.y * scalar)
    
    def dot(self, other):
        return self.x * other.x + self.y * other.y
    
    def cross(self, other):
        return self.x * other.y - self.y * other.x
    
    def length(self):
        return np.sqrt(self.dot(self))

class GJK:
    """Gilbert-Johnson-Keerthi碰撞检测算法"""
    def __init__(self, max_iterations=30):
        self.max_iterations = max_iterations
    
    def _support(self, direction, vertices):
        """计算支撑点：返回在给定方向上最远的顶点"""
        return max(vertices, key=lambda v: v.dot(direction))
    
    def _get_vertices(self, square):
        """获取方块的顶点"""
        vertices = [
            Vector2D(square.x, square.y),  # 左下
            Vector2D(square.x + square.width, square.y),  # 右下
            Vector2D(square.x + square.width, square.y + square.height),  # 右上
            Vector2D(square.x, square.y + square.height)  # 左上
        ]
        return vertices
    
    def _minkowski_difference(self, direction, square1, square2):
        """计算两个方块在给定方向上的闵可夫斯基差集的支撑点"""
        vertices1 = self._get_vertices(square1)
        vertices2 = self._get_vertices(square2)
        
        support1 = self._support(direction, vertices1)
        support2 = self._support(direction * -1, vertices2)
        return support1 - support2
    
    def check_collision(self, square1, square2):
        """检查两个方块是否碰撞"""
        # 初始方向
        direction = Vector2D(1, 0)
        
        # 获取第一个单纯形顶点
        simplex = [self._minkowski_difference(direction, square1, square2)]
        
        # 更新方向朝向原点
        direction = Vector2D(0, 0) - simplex[0]
        
        for _ in range(self.max_iterations):
            # 获取新的支撑点
            support = self._minkowski_difference(direction, square1, square2)
            
            # 如果新点没有在方向上更接近原点，则无碰撞
            if support.dot(direction) <= 0:
                return False
            
            # 添加新点到单纯形
            simplex.append(support)
            
            # 更新单纯形和搜索方向
            if self._do_simplex(simplex, direction):
                return True
        
        return False
    
    def _do_simplex(self, simplex, direction):
        """处理单纯形，更新搜索方向"""
        if len(simplex) == 2:
            return self._do_line(simplex, direction)
        elif len(simplex) == 3:
            return self._do_triangle(simplex, direction)
        return False
    
    def _do_line(self, simplex, direction):
        """处理线段情况"""
        B = simplex[0]
        A = simplex[1]
        AB = B - A
        AO = Vector2D(0, 0) - A
        
        # 计算垂直于AB的方向
        ABperp = Vector2D(-AB.y, AB.x)
        if ABperp.dot(AO) > 0:
            direction.x = ABperp.x
            direction.y = ABperp.y
        else:
            direction.x = -ABperp.x
            direction.y = -ABperp.y
        return False
    
    def _do_triangle(self, simplex, direction):
        """处理三角形情况"""
        C = simplex[0]
        B = simplex[1]
        A = simplex[2]
        
        AB = B - A
        AC = C - A
        AO = Vector2D(0, 0) - A
        
        # 计算AB的法向量
        ABperp = Vector2D(-AB.y, AB.x)
        # 计算AC的法向量
        ACperp = Vector2D(AC.y, -AC.x)
        
        if ABperp.dot(AC) > 0:
            if AB.cross(AO) > 0:
                simplex.remove(C)
                direction.x = ABperp.x
                direction.y = ABperp.y
                return False
            if AO.cross(AC) > 0:
                simplex.remove(B)
                direction.x = ACperp.x
                direction.y = ACperp.y
                return False
            return True
        
        if AC.cross(AO) > 0:
            simplex.remove(B)
            direction.x = ACperp.x
            direction.y = ACperp.y
            return False
        if AO.cross(AB) > 0:
            simplex.remove(C)
            direction.x = ABperp.x
            direction.y = ABperp.y
            return False
        return True

def draw_gjk_collision(squares):
    """使用GJK算法绘制碰撞检测结果"""
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    
    gjk = GJK()
    start_time = time.time()
    
    # 检测碰撞
    collisions = set()
    for i, square1 in enumerate(squares):
        for j, square2 in enumerate(squares[i+1:], i+1):
            if gjk.check_collision(square1, square2):
                collisions.add(tuple(sorted([i, j])))
    
    detection_time = time.time() - start_time
    
    # 绘制方块
    for i, square in enumerate(squares):
        is_colliding = any(i in collision for collision in collisions)
        color = 'red' if is_colliding else 'blue'
        
        # 绘制方块填充
        rect = MplRectangle((square.x, square.y), 1, 1,
                          facecolor=color, alpha=0.5)
        ax.add_patch(rect)
        
        # 绘制顶点
        vertices = [(square.x, square.y),
                   (square.x + 1, square.y),
                   (square.x + 1, square.y + 1),
                   (square.x, square.y + 1)]
        vertices.append(vertices[0])  # 闭合多边形
        xs, ys = zip(*vertices)
        ax.plot(xs, ys, color='black', linewidth=1, alpha=0.8)
        
        # 标注编号
        ax.text(square.x + 0.5, square.y + 0.5, str(i+1),
                ha='center', va='center',
                color='white', fontweight='bold')
    
    # 绘制碰撞连线
    for i, j in collisions:
        s1, s2 = squares[i], squares[j]
        center1 = (s1.x + 0.5, s1.y + 0.5)
        center2 = (s2.x + 0.5, s2.y + 0.5)
        ax.plot([center1[0], center2[0]], [center1[1], center2[1]],
                'y--', alpha=0.5)
    
    ax.set_xlim(-1, 11)
    ax.set_ylim(-1, 11)
    ax.grid(True, linestyle='--', alpha=0.3)
    ax.set_title(f'GJK Collision Detection\nDetection Time: {detection_time:.4f}s\n'
                f'Collisions Found: {len(collisions)} pairs')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nGJK算法检测结果:")
    print(f"- 检测时间: {detection_time:.4f}秒")
    print(f"- 检测到的碰撞对数: {len(collisions)}")
    if collisions:
        print("- 碰撞对:", sorted(collisions))

# 运行GJK碰撞检测
print("GJK算法碰撞检测：")
draw_gjk_collision(squares)
```

```python
def draw_gjk_collision_detailed(squares):
    """使用GJK算法绘制详细的碰撞检测结果"""
    fig = plt.figure(figsize=(15, 12))
    ax = fig.add_subplot(111)
    
    gjk = GJK()
    start_time = time.time()
    
    # 检测碰撞
    collisions = []
    collision_processes = []  # 存储碰撞检测的过程数据
    
    for i, square1 in enumerate(squares):
        for j, square2 in enumerate(squares[i+1:], i+1):
            # 记录GJK迭代过程
            simplex_history = []
            direction = Vector2D(1, 0)
            
            # 获取第一个单纯形顶点
            first_support = gjk._minkowski_difference(direction, square1, square2)
            simplex = [first_support]
            simplex_history.append((simplex[:], direction))
            
            # 更新方向朝向原点
            direction = Vector2D(0, 0) - first_support
            
            is_colliding = False
            for _ in range(gjk.max_iterations):
                support = gjk._minkowski_difference(direction, square1, square2)
                if support.dot(direction) <= 0:
                    break
                
                simplex.append(support)
                simplex_history.append((simplex[:], direction))
                
                if gjk._do_simplex(simplex, direction):
                    is_colliding = True
                    collisions.append((i, j))
                    break
            
            if is_colliding:
                collision_processes.append((i, j, simplex_history))
    
    detection_time = time.time() - start_time
    
    # 绘制方块
    for i, square in enumerate(squares):
        is_colliding = any(i in collision for collision in collisions)
        color = 'red' if is_colliding else 'blue'
        
        # 绘制方块填充
        rect = MplRectangle((square.x, square.y), 1, 1,
                          facecolor=color, alpha=0.5,
                          edgecolor='black', linewidth=1.5)
        ax.add_patch(rect)
        
        # 绘制顶点和边框
        vertices = [(square.x, square.y),
                   (square.x + 1, square.y),
                   (square.x + 1, square.y + 1),
                   (square.x, square.y + 1)]
        vertices.append(vertices[0])
        xs, ys = zip(*vertices)
        ax.plot(xs, ys, color='black', linewidth=1.5, alpha=0.8)
        
        # 标注编号
        ax.text(square.x + 0.5, square.y + 0.5, str(i+1),
                ha='center', va='center',
                color='white', fontweight='bold', fontsize=10)
    
    # 绘制碰撞连线和过程
    for i, j, history in collision_processes:
        s1, s2 = squares[i], squares[j]
        center1 = (s1.x + 0.5, s1.y + 0.5)
        center2 = (s2.x + 0.5, s2.y + 0.5)
        
        # 绘制碰撞连线
        ax.plot([center1[0], center2[0]], [center1[1], center2[1]],
                color='orange', linewidth=2, alpha=0.8,
                linestyle='--', label=f'Collision {i+1}-{j+1}')
        
        # 绘制GJK迭代过程
        colors = plt.cm.rainbow(np.linspace(0, 1, len(history)))
        for idx, ((simplex, direction), color) in enumerate(zip(history, colors)):
            # 绘制单纯形
            if len(simplex) >= 2:
                simplex_points = [(p.x, p.y) for p in simplex]
                simplex_points.append(simplex_points[0])
                xs, ys = zip(*simplex_points)
                ax.plot(xs, ys, color=color, alpha=0.3, linewidth=1,
                       label=f'Iteration {idx+1}' if idx == 0 else "")
    
    # 设置图形属性
    ax.set_xlim(-1, 11)
    ax.set_ylim(-1, 11)
    ax.grid(True, linestyle='--', alpha=0.3)
    ax.set_title(f'GJK Collision Detection (Detailed)\n'
                f'Detection Time: {detection_time:.4f}s\n'
                f'Collisions Found: {len(collisions)} pairs')
    
    # 添加图例
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.tight_layout()
    plt.show()
    
    # 打印详细信息
    print(f"\nGJK算法检测结果:")
    print(f"- 检测时间: {detection_time:.4f}秒")
    print(f"- 检测到的碰撞对数: {len(collisions)}")
    if collisions:
        print("- 碰撞对:", sorted(collisions))
        print("\n碰撞检测过程:")
        for i, j, history in collision_processes:
            print(f"\n方块 {i+1} 和方块 {j+1} 的碰撞检测:")
            print(f"- 迭代次数: {len(history)}")
            print(f"- 最终单纯形顶点数: {len(history[-1][0])}")

# 运行详细的GJK碰撞检测可视化
print("GJK算法碰撞检测（详细版本）：")
draw_gjk_collision_detailed(squares)
```
